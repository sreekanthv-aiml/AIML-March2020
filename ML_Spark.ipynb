```python
from pyspark.sql.functions import col
dataset = df.select (col('Survived').cast('float'),
                    col('Pclass').cast('float'),
                    col('Sex'),
                    col('Age').cast('float'),
                    col('Fare').cast('float'),
                    col('Embarked'))
```

```python
from pyspark.sql.functions import isnull, when , count , col
dataset.select ([count(when(isnull(c),1)).alias(c) for c in dataset.columns]).show()
```

```python
from pyspark.sql.functions import isnull, when , count , col
dataset.select ([count(when(isnull(col('Age')),1)).alias('Age') ]).show()
```

```python
dataset =dataset.replace('?' , None)\
    .dropna(how='any')
```

```python
dataset
```

```python
dataset.count()
```

```python
from pyspark.ml.feature import StringIndexer
dataset = StringIndexer(inputCol='Sex',
                       outputCol='Gender',
                       handleInvalid='keep').fit(dataset).transform(dataset)
```

```python
dataset
```

```python
dataset.show()
```

```python
dataset = StringIndexer(inputCol='Embarked',
                       outputCol='Boarded',
                       handleInvalid='keep').fit(dataset).transform(dataset)
```

```python
dataset.show()
```

```python
dataset=dataset.drop('Sex')
dataset=dataset.drop('Embarked')
dataset.show()
```

```python
required_features=['Pclass','Age','Fare','Gender','Boarded']
```

```python
from pyspark.ml.feature import VectorAssembler
assembler = VectorAssembler(inputCols=required_features,outputCol='features')
transformed_data= assembler.transform(dataset)
```

```python
transformed_data
```

```python
transformed_data.show()
```

```python
(training_data,test_data)= transformed_data.randomSplit([0.8,0.2])
```

```python
training_data.show()
```

```python
test_data.show()
```

```python
from pyspark.ml.classification import RandomForestClassifier
rf=RandomForestClassifier(labelCol='Survived',
                         featuresCol='features',
                         maxDepth=5)
```

```python
model = rf.fit(training_data)
```

```python
model
```

```python
pred =model.transform(test_data)
```

```python
pred.show()
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
evaluator =MulticlassClassificationEvaluator (labelCol='Survived', predictionCol='prediction',metricName='accuracy')
```

```python
accuracy= evaluator.evaluate (pred)
print('Test Accuracy=' , accuracy)
```
